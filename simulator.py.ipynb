```python
num_iterations = 10000

extinction1_before = []
extinction2_before = []
extinction1_after = []
extinction2_after = []
pop_grows = []
pop_shrink = []
pop_split = []
eq_point = []
#TODO: Current High score

for i in range(num_iterations):
    k1 = random.uniform() * 25
    k2 = random.uniform() * 25
    l1 = random.uniform() * 25
    l2 = random.uniform() * 25
    c1 = random.uniform() * 25
    c2 = random.uniform() * 25
    bomb_time = random.uniform() * 100
    x = random.uniform() * 100
    y = random.uniform() * 50 + 25
    dt = 0.001; max_time = 100

    # initial time and populations
    t = 0

    # empty lists in which to store time and populations
    t_list = []; x_list = []; y_list = []

    # initialize lists
    t_list.append(t); x_list.append(x); y_list.append(y)

    time_bomb_armed = True

    pop_max_x_before = 0
    pop_max_y_before = 0
    pop_max_x_after = 0
    pop_max_y_after = 0

    iteration = {"k1":k1, "k2":k2, "l1":l1, "l2":l2, "c1":c1,
                    "c2":c2, "bt": bomb_time, "x":x, "y": y}
    check = True

    while t < max_time:
        # calc new values for t, x, y
        t = t + dt
        x = x + (k1*x*(1-x/l1) - c1*x*y)*dt
        y = y + (k2*y*(1-y/l2) - c2*x*y)*dt

        if time_bomb_armed and t>bomb_time:

            if x == 0 and y == 0:
                extinction2_before.append(iteration)
                check = False
            elif x == 0 or y == 0:
                extinction1_before.append(iteration)
                check = False

            x = 0.5 * x
            y = 0.5 * y
            time_bomb_armed = False

        # store new values in lists
        t_list.append(t)
        x_list.append(x)
        y_list.append(y)

        if x > pop_max_x_before and t < bomb_time:
            pop_max_x_before = x
        if y > pop_max_y_before and t < bomb_time:
            pop_max_y_before = y

        if x > pop_max_x_after and t > bomb_time:
                pop_max_x_after = x
        if y > pop_max_y_after and t > bomb_time:
            pop_max_y_after = y

    if check:
        check_x = pop_max_x_before > pop_max_x_after
        check_y = pop_max_y_before > pop_max_y_after
        if x == 0 and y == 0:
            extinction2_after.append(iteration)
        elif x == 0 or y == 0:
            extinction1_after.append(iteration)
        elif pop_max_x_before == pop_max_x_after and pop_max_y_before == pop_max_y_after:
            eq_point.append(iteration)
        elif check_x and check_y:
            pop_shrink.append(iteration)
        elif not check_x and not check_y:
            pop_grows.append(iteration)
        elif check_x or check_y:
            pop_split.append(iteration)

with open("competing_species_e1b.txt", "w") as f:
    json.dump(extinction1_before, f)
with open("competing_species_e2b.txt", "w") as f:
    json.dump(extinction2_before, f)
with open("competing_species_e1a.txt", "w") as f:
    json.dump(extinction1_after, f)
with open("competing_species_e2a.txt", "w") as f:
    json.dump(extinction2_after, f)
with open("competing_species_eq.txt", "w") as f:
    json.dump(eq_point, f)
with open("competing_species_shrink.txt", "w") as f:
    json.dump(pop_shrink, f)
with open("competing_species_grow.txt", "w") as f:
    json.dump(pop_grows, f)
with open("competing_species_split.txt", "w") as f:
    json.dump(pop_split, f)
with open("competing_species_results.txt", "w") as f:
    res = """e1b: {:.2f}, e2b: {:.2f}, e1a: {:.2f}, e2a: {:.2f},eq: {:.2f}, shrink: {:.2f}, grow: {:.2f}, split: {:.2f}, """.format(
        len(extinction1_before)/num_iterations, len(extinction2_before)/num_iterations,
        len(extinction1_after)/num_iterations, len(extinction2_after)/num_iterations,
        len(eq_point)/num_iterations, len(pop_shrink)/num_iterations,
        len(pop_grows)/num_iterations, len(pop_split)/num_iterations)
    json.dump(res, f)
```
